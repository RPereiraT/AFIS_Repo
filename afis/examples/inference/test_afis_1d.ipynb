# A_FIS Test - 1D Case

Test A_FIS with 1-dimensional input domain.


````python
# Setup path and imports
import sys
from pathlib import Path

# Add AFIS_Repo root to path
REPO_ROOT = Path.cwd().parent.parent.parent
sys.path.insert(0, str(REPO_ROOT))

import numpy as np

# Import from afis package
from afis.core.A_FIS import format_FN_N_Dim
from afis.core.afis_utils import FuzzySet, Trapezoidal
from afis.visualization.plotting import (
    create_rule_base, run_afis, plot_results, show_svi_table,
    show_detailed_diagnostic, compute_activation_curves,
    plot_activation_curves
)

print("Modules loaded successfully!")
````

## Create Rule Base


````python
# Define membership functions (antecedents)
A_1 = Trapezoidal(ini=10, top1=20, top2=20, end=35)
A_2 = Trapezoidal(ini=40, top1=50, top2=65, end=70)
A_3 = Trapezoidal(ini=95, top1=110, top2=110, end=120)
A_4 = Trapezoidal(ini=120, top1=140, top2=140, end=140)

# Define membership functions (consequents)
B_1 = Trapezoidal(ini=0, top1=0, top2=0, end=100)
B_2 = Trapezoidal(ini=0, top1=100, top2=100, end=200)
B_3 = Trapezoidal(ini=100, top1=200, top2=200, end=300)
B_4 = Trapezoidal(ini=200, top1=300, top2=300, end=300)

# Create FuzzySet objects
A_1_fs = FuzzySet("A_1", A_1)
A_2_fs = FuzzySet("A_2", A_2)
A_3_fs = FuzzySet("A_3", A_3)
A_4_fs = FuzzySet("A_4", A_4)

B_1_fs = FuzzySet("B_1", B_1)
B_2_fs = FuzzySet("B_2", B_2)
B_3_fs = FuzzySet("B_3", B_3)
B_4_fs = FuzzySet("B_4", B_4)

# Create rule base
rule_base_1d = create_rule_base(
    antecedents=[[A_1_fs, A_2_fs, A_3_fs, A_4_fs]],
    consequents=[B_1_fs, B_2_fs, B_3_fs, B_4_fs],
    input_ranges=[(0, 150)],
    output_range=(0, 300),
    rules=[([0], 0), ([1], 1), ([2], 3), ([3], 2)]
)

print("Rule base created with 4 rules:")
print("  Rule 0: A_1 → B_1")
print("  Rule 1: A_2 → B_2")
print("  Rule 2: A_3 → B_3")
print("  Rule 3: A_4 → B_4")
````

## Define Input and Execute


````python
# Define input
input1 = Trapezoidal(ini=80, top1=80, top2=80, end=80)  # Crisp input at 80
input1 = Trapezoidal(ini=49, top1=49, top2=49, end=49)  # Crisp input at 80
input1 = Trapezoidal(ini=90, top1=95, top2=95, end=100)  # Crisp input at 80

input_list_1d = [input1]

# Format inputs
input_fuzzy_sets_1d = [FuzzySet(f"Input", inp) for inp in input_list_1d]
input_formatted_1d = format_FN_N_Dim(input_fuzzy_sets_1d, ftype='rule_antecedent')

print("Input value:")
print(f"  {input_formatted_1d[0]}")

# Run A_FIS
output_1d, U_1d, y_crisp_1d, max_values_1d = run_afis(
    input_formatted_1d, 
    rule_base_1d,
    imp_params=['luka', 1], 
    t_norm_type='product'
)

print(f"\nDefuzzified output: {y_crisp_1d:.3f}")

# Plot results (shows antecedents, inputs, and output)
plot_results(rule_base_1d, input_list_1d, output_1d, U_1d, y_crisp_1d)

# Show S_A table
show_svi_table(rule_base_1d, input_formatted_1d)
````

## Detailed Diagnostic


````python
# Detailed step-by-step diagnostic (works for both 1D and ND!)
show_detailed_diagnostic(rule_base_1d, input_formatted_1d)
````

## Activation Curves (S_A vs Input Position)

````python
# Functions are now imported from test_utils:
# - compute_activation_curves(rule_base, dim, base_width=0, step=1, imp_params=None)
# - plot_activation_curves(rule_base, dim, C_i_values, S_A_curves, antecedent_order)
#
# Note: imp_params defaults to rule_base._last_imp_params (from last run_afis call)

print("Activation curve functions imported from test_utils")
````

````python
# Example: Compute and plot activation curves for the 1D case

# Parameters
base_width = 5  # Base width of isosceles triangle input (0 for crisp/singleton)
step = 0.5        # Displacement step

# Compute activation curves for dimension 0
C_i_values, S_A_curves, antecedent_order = compute_activation_curves(
    rule_base_1d, 
    dim=0, 
    base_width=base_width, 
    step=step
)

# Plot
fig = plot_activation_curves(rule_base_1d, dim=0, C_i_values=C_i_values, 
                             S_A_curves=S_A_curves, antecedent_order=antecedent_order)
fig.show()
````

### Usage for 2D Case
For a 2D rule base, call the functions for each dimension:

```python
# Dimension 0
C_i_0, S_A_0, order_0 = compute_activation_curves(rule_base_2d, dim=0, base_width=5, step=1)
fig0 = plot_activation_curves(rule_base_2d, dim=0, C_i_0, S_A_0, order_0)
fig0.show()

# Dimension 1
C_i_1, S_A_1, order_1 = compute_activation_curves(rule_base_2d, dim=1, base_width=5, step=1)
fig1 = plot_activation_curves(rule_base_2d, dim=1, C_i_1, S_A_1, order_1)
fig1.show()
```

**Note on operators:** By default, `compute_activation_curves` uses the same `imp_params` from your last `run_afis()` call (stored in `rule_base._last_imp_params`). You can also pass them explicitly:

```python
C_i, S_A, order = compute_activation_curves(rule_base, dim=0, base_width=5, imp_params=['luka', 1])
```
